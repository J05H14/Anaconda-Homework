PART A. Importing the Needed Modules from in-class tutorial

```python
import numpy as np

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

from keras.layers import Dense, Dropout, Activation, Flatten
from keras.models import Sequential
from keras.utils import np_utils

from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
```

PART B. Import mnist and split and reshape

```python
from keras.datasets import mnist

#loading the data from mnist and reshaping it
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

PART C. OneHotEncoding on label y

```python
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
print (y_train.shape)
print (y_train[:10,:])
```

PART D. Define model_creator function

```python
input_size = 784
hidden_neurons = 100
out_size = 10
def model_creator():
   
    # define:
    model = Sequential()
    # design the structure:
    model.add(Dense(hidden_neurons, input_dim = input_size))
    model.add(Activation('sigmoid')) # Activation
    
    model.add(Dense(out_size, input_dim = hidden_neurons))  # Nuerons
    model.add(Activation('softmax')) # Activation
    # compile:
    model.compile(loss='categorical_crossentropy', optimizer='adam',
    metrics=['accuracy'])
    # return:
    return model 
```

PART E. Fixing the random state

```python
seed = 2
np.random.seed(seed)
```

PART F. Wrapmodel as an Object using KerasClassifier

```python
model = KerasClassifier(build_fn=model_creator,verbose=2)
```

PART G. Using sklearn GridSearch

```python
# batch_size = [30,50,100]
# epochs = [10,15,20]
```

```python
epochs = np.array([10 , 15 , 20])
batches = np.array([30 , 50 , 100])
param_grid = dict(epochs=epochs, batch_size=batches)
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_log_loss', cv=10)
```

```python
grid_result = grid.fit(X_train, y_train, verbose=1)
```

```python
#finding out the best batch_size and best number of epschs
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

Best Batch Size and Epochs is 50 and 20 respectively

Obtain the accuracy from the Best estimate against the training set

```python
score = grid.best_estimator_.model.evaluate(X_test, y_test)
```

Printing out the score

```python
#score[0] would be the neg log loss
print("Score: " + str(score[1]))
```
